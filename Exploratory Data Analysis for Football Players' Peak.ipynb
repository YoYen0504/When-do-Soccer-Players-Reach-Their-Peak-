```python
import pandas as pd
```

```python
data_clean_df = pd.read_csv("data_clean.csv")
```

```python
data_clean_df.head()
```

```python
data_clean_df.drop(columns = "Unnamed: 0", axis = 1, inplace = True)
```

```python
### 開始計算球員在巔峰時的年齡，先做資料清理

### 先將生日為空值的資料去除
drop_no_age_index = data_clean_df[data_clean_df["Date of birth"].str.count(" ") == 1].index
data_clean_df.drop(drop_no_age_index, inplace = True)

### 將最高身價日期未知的資料去除
drop_no_highest_market_value_date_index = data_clean_df[data_clean_df["Highest market value date"].str.count("-") == 1].index
data_clean_df.drop(drop_no_highest_market_value_date_index, inplace = True)
```

```python
### 開始計算球員在巔峰時的年齡

from datetime import datetime
player_age_at_highest_market_value = []

for num_of_rows in range(data_clean_df.shape[0]):
    birth_date = data_clean_df.iloc[num_of_rows, 1].strip().replace(" Happy Birthday", "")
    
    highest_market_value_date = data_clean_df.iloc[num_of_rows, -1].strip()
    
    dateFormatter = "%b %d, %Y"
    birth_datetime = datetime.strptime(birth_date, dateFormatter)
    highest_market_value_datetime = datetime.strptime(highest_market_value_date, dateFormatter)
    
    days_diff = float(str(highest_market_value_datetime - birth_datetime).replace(" days, 0:00:00", ""))
    player_age_at_highest_market_value_datetime = days_diff/365
    
    player_age_at_highest_market_value.append(round(player_age_at_highest_market_value_datetime, 2))
```

```python
### 將計算出來的最高身價時的年齡併入DataFrame中

highest_market_value_age = pd.DataFrame(player_age_at_highest_market_value) # 將儲存最高身價時的年齡的list轉成DataFrame

data_clean_df.reset_index(inplace = True) # 重置資料的index，DataFrame合併時會將index一樣的資料併在一起
data_clean_df.drop(columns = "index", axis = 1, inplace = True) # 將舊的index刪除

concat_data_clean_df = pd.concat([data_clean_df, highest_market_value_age], axis = 1) # 將原本的df與最高身價時的年齡的df合併
concat_data_clean_df.rename({0:"Highest market value age"}, axis = 1, inplace = True) # 更改"最高身價時的年齡"的column名稱
```

```python
concat_data_clean_df
```

```python
final_df = concat_data_clean_df.copy()
```

```python
### 去除 Highest market value date 為 Oct 4, 2004 的資料
highest_market_value_date_oct_4_2004 = final_df[final_df["Highest market value date"].str.count("Oct 4, 2004") == 1].index
final_df.drop(highest_market_value_date_oct_4_2004, inplace = True)

### 去除正值顛峰的資料( abs(Highest market value age - Age) < 1.0 的資料)
best_shape_data_index = final_df[abs(final_df["Age"] - final_df["Highest market value age"]) < 1 ].index
final_df.drop(best_shape_data_index, inplace = True)

### 看一下球員在巔峰時的平均年齡是多少
final_df["Highest market value age"].mean()
```
